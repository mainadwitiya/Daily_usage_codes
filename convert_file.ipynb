```python
import glob
import os, sys
from PIL import Image
save_path='/home/daisy/Documents/final_images/'
if not os.path.exists(save_path):
    os.mkdir(save_path)

files = glob.glob('/home/daisy/Documents/IMAGES/Images/*')
for file in files:
    if file.endswith('.tif'):
        infile=os.path.basename(file)
        #print(infile[:-4])
        im = Image.open(file)
        outfile = infile[:-4] + ".jpg"
        out = im.convert("RGB")
        out.save(save_path+outfile, "JPEG", quality=90)
        #print('yes')
    elif file.endswith('.jpg'):
        shutil.copy(file,save_path)
        #print('no')
```
